```python
import tkinter as tk
from tkinter import ttk
from tkinter import messagebox
import csv
import os

color_file_path = 'data/colors.csv'
brand_file_path = 'data/brands.csv'
clothes_file_path = 'data/clothes.csv'
clothing_photos_folder = 'clothing_photos/'

types_list = ['Top', 'Bottom', 'Shoes', 'Accessories', 'Jewelry', "Outerwear", "Bag", "Dress"]
top_list = ['Sweater', 'Cardi', 'Tank', 'Tee', 'Long-Sleeve', 'Bodysuit', 'Sweatshirt']
bottom_list = ['Jeans', 'Trousers', 'Mini Skirt', 'Midi Skirt', 'Maxi Skirt', 'Shorts', 'Sweatpants']
shoes_list = ['Heels', 'Boots', 'Sneakers', 'Slippers', 'Sandals']
accessories_list = ['Hair', ] 
jewelry_list = ['Necklace', 'Bracelet', 'Anklet', 'Earrings', 'Ring']
outerwear_list = ['Jacket', 'Coat', 'Blazer']
bag_list = ['Backpack', 'Purse', 'Tote']
dress_list = ['Mini', 'Midi', 'Maxi']

def add_color():
    clear_form()
    label = tk.Label(window, text="Enter Color:")
    label.pack(padx=10, pady=10)
    
    color_entry = tk.Entry(window, width=20)
    color_entry.pack(padx=10, pady=10)
    
    add_button = tk.Button(window, text="Add Color", command=lambda: save_color(color_entry))
    add_button.pack(padx=10, pady=10)

def add_brand():
    clear_form()
    label = tk.Label(window, text="Enter Brand:")
    label.pack(padx=10, pady=10)
    
    brand_entry = tk.Entry(window, width=20)
    brand_entry.pack(padx=10, pady=10)
    
    add_button = tk.Button(window, text="Add Brand", command=lambda: save_brand(brand_entry))
    add_button.pack(padx=10, pady=10)

def add_clothes():
    clear_form()

    name_label = tk.Label(window, text="Enter Name:")
    name_label.pack(padx=10, pady=10)
    
    name_entry = tk.Entry(window, width=20)
    name_entry.pack(padx=10, pady=10)

    type_label = tk.Label(window, text="Select Type:")
    type_label.pack(padx=10, pady=10)

    type_dropdown = ttk.Combobox(window, values=types_list)
    type_dropdown.pack(padx=10, pady=10)

    subtype_label = tk.Label(window, text="Select Subtype:")
    subtype_label.pack(padx=10, pady=10)

    subtype_dropdown = ttk.Combobox(window, values=[])  # Initially empty
    subtype_dropdown.pack(padx=10, pady=10)

    # Update subtypes when type changes
    def update_subtypes(event):
        selected_type = type_dropdown.get()
        if selected_type == "Top":
            subtype_dropdown['values'] = top_list
        elif selected_type == "Bottom":
            subtype_dropdown['values'] = bottom_list
        elif selected_type == "Shoes":
            subtype_dropdown['values'] = shoes_list
        elif selected_type == "Accessories":
            subtype_dropdown['values'] = accessories_list
        elif selected_type == "Jewelry":
            subtype_dropdown['values'] = jewelry_list
        elif selected_type == "Outerwear":
            subtype_dropdown['values'] = outerwear_list
        elif selected_type == "Bag":
            subtype_dropdown['values'] = bag_list
        elif selected_type == "Dress":
            subtype_dropdown['values'] = dress_list
        else:
            subtype_dropdown['values'] = []  # Clear if no match

    type_dropdown.bind("<<ComboboxSelected>>", update_subtypes)

    brand_label = tk.Label(window, text="Select Brand:")
    brand_label.pack(padx=10, pady=10)

    brand_dropdown = ttk.Combobox(window, values=get_brands())
    brand_dropdown.pack(padx=10, pady=10)

    photo_label = tk.Label(window, text="Enter Photo Name (e.g., navy_sweats):")
    photo_label.pack(padx=10, pady=10)

    photo_entry = tk.Entry(window, width=20)
    photo_entry.pack(padx=10, pady=10)

    color_label = tk.Label(window, text="Select Colors:")
    color_label.pack(padx=10, pady=10)

    color_dropdown = tk.Listbox(window, selectmode=tk.MULTIPLE, height=6, width=20)
    with open(color_file_path, mode='r') as file:
        reader = csv.reader(file)
        colors = [row[0] for row in reader]
    for color in colors:
        color_dropdown.insert(tk.END, color)
    color_dropdown.pack(padx=10, pady=10)

    add_button = tk.Button(window, text="Add Clothes", 
                           command=lambda: save_clothes(
                               name_entry, type_dropdown, subtype_dropdown, 
                               color_dropdown, brand_dropdown, photo_entry))
    add_button.pack(padx=10, pady=10)

def add_outfit():
    clear_form()

    date_label = tk.Label(window, text="Enter Date (MM/DD/YYYY):")
    date_label.pack(padx=10, pady=10)
    
    date_entry = tk.Entry(window, width=20)
    date_entry.pack(padx=10, pady=10)

    photo_label = tk.Label(window, text="Enter Photo Name (e.g., 01.01.2025):")
    photo_label.pack(padx=10, pady=10)
    
    photo_entry = tk.Entry(window, width=20)
    photo_entry.pack(padx=10, pady=10)

    items_label = tk.Label(window, text="Select Clothes:")
    items_label.pack(padx=10, pady=10)

    items_dropdown = tk.Listbox(window, selectmode=tk.MULTIPLE, height=6, width=20)
    
    with open(clothes_file_path, mode='r') as file:
        reader = csv.reader(file)
        header = next(reader)
        items = [row[0] for row in reader]

    for item in items:
        items_dropdown.insert(tk.END, item)
    items_dropdown.pack(padx=10, pady=10)

    add_button = tk.Button(window, text="Add Outfit", command=lambda: save_outfit(date_entry, photo_entry, items_dropdown))
    add_button.pack(padx=10, pady=10)

def save_color(color_entry):
    color = color_entry.get().strip()
    if color:
        with open(color_file_path, mode='r') as file:
            reader = csv.reader(file)
            existing_colors = [row[0] for row in reader]
            if color in existing_colors:
                messagebox.showerror("Duplicate Color", f"The color '{color}' already exists.")
                return

        with open(color_file_path, mode='a', newline='') as file:
            writer = csv.writer(file)
            writer.writerow([color])
        color_entry.delete(0, tk.END)
    
    show_main_menu()

def save_brand(brand_entry):
    brand = brand_entry.get().strip()
    if brand:
        with open(brand_file_path, mode='r') as file:
            reader = csv.reader(file)
            existing_brands = [row[0] for row in reader]
            if brand in existing_brands:
                messagebox.showerror("Duplicate Brand", f"The brand '{brand}' already exists.")
                return

        with open(brand_file_path, mode='a', newline='') as file:
            writer = csv.writer(file)
            writer.writerow([brand])
        brand_entry.delete(0, tk.END)
    
    show_main_menu()

def get_brands():
    with open(brand_file_path, mode='r') as file:
        reader = csv.reader(file)
        brands = [row[0] for row in reader]
    return brands

def save_clothes(name_entry, type_dropdown, subtype_dropdown, color_dropdown, brand_dropdown, photo_entry):
    name = name_entry.get().strip()
    type_ = type_dropdown.get().strip()
    subtype = subtype_dropdown.get().strip()
    selected_colors = [color_dropdown.get(i) for i in color_dropdown.curselection()]
    colors = ", ".join(selected_colors)
    brand = brand_dropdown.get().strip()
    photo_name = photo_entry.get().strip()
    photo_path = f"{clothing_photos_folder}{photo_name}.jpg"

    if not name or not type_ or not subtype or not colors or not brand or not photo_name:
        messagebox.showerror("Missing Fields", "Please fill in all fields.")
        return

    if not os.path.exists(photo_path):
        messagebox.showerror("File Not Found", f"The photo '{photo_name}.jpg' was not found in the '{clothing_photos_folder}' folder.")
        return

    with open(clothes_file_path, mode='r') as file:
        reader = csv.reader(file)
        header = next(reader)
        existing_names = [row[0] for row in reader]
        if name in existing_names:
            messagebox.showerror("Duplicate Name", f"The clothing item '{name}' already exists.")
            return

    with open(clothes_file_path, mode='a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow([name, type_, subtype, colors, brand, 0, photo_path])

    name_entry.delete(0, tk.END)
    type_dropdown.set("")
    subtype_dropdown.set("")
    color_dropdown.selection_clear(0, tk.END)
    brand_dropdown.set("")
    photo_entry.delete(0, tk.END)

    show_main_menu()

def save_outfit(date_entry, photo_entry, items_dropdown):
    date = date_entry.get().strip()
    photo_name = photo_entry.get().strip()
    items = [items_dropdown.get(i) for i in items_dropdown.curselection()]

    if not date or not photo_name or not items:
        messagebox.showerror("Missing Fields", "Please fill in all fields.")
        return

    photo_path = f"outfit_photos/{photo_name}.jpg"

    if not os.path.exists(photo_path):
        messagebox.showerror("File Not Found", f"The photo '{photo_name}.jpg' was not found in the 'outfit_photos' folder.")
        return

    with open('data/outfits.csv', mode='r') as file:
        reader = csv.reader(file)
        for row in reader:
            if row and row[1] == photo_path:
                messagebox.showerror("Duplicate Photo", f"The photo '{photo_name}.jpg' already exists in the outfits database.")
                return

    # Increment the counter for selected clothing items
    clothes_data = []
    with open(clothes_file_path, mode='r') as file:
        reader = csv.reader(file)
        header = next(reader)
        clothes_data.append(header)
        for row in reader:
            if row[0] in items:
                try:
                    # Ensure row[5] (Counter) is numeric before incrementing
                    row[5] = str(int(row[5]) + 1)
                except ValueError:
                    messagebox.showwarning("Data Error", f"Invalid counter value in row: {row}. Skipping this item.")
                    continue
            clothes_data.append(row)

    with open(clothes_file_path, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerows(clothes_data)

    # Save the outfit to outfits.csv
    with open('data/outfits.csv', mode='a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow([date, photo_path, ", ".join(items), 0])  # OOTW initialized to 0

    date_entry.delete(0, tk.END)
    photo_entry.delete(0, tk.END)
    items_dropdown.selection_clear(0, tk.END)

    show_main_menu()

def clear_form():
    for widget in window.winfo_children():
        widget.pack_forget()

def show_main_menu():
    clear_form()

    add_color_button = tk.Button(window, text="Add Color", command=add_color)
    add_color_button.pack(padx=10, pady=10)

    add_brand_button = tk.Button(window, text="Add Brand", command=add_brand)
    add_brand_button.pack(padx=10, pady=10)

    add_clothes_button = tk.Button(window, text="Add Clothes", command=add_clothes)
    add_clothes_button.pack(padx=10, pady=10)

    add_outfit_button = tk.Button(window, text="Add Outfit", command=add_outfit)
    add_outfit_button.pack(padx=10, pady=10)

window = tk.Tk()
window.title("365 Days of Style")

show_main_menu()

window.mainloop()
```
